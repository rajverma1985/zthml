```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/car-sales.csv")
```

```python
df['Price'] = df['Price'].str.replace('[\$\.\,]', '', regex=True).astype(int)
df['Price'] = df['Price']/100
```

```python
# saving the cleaned data to csv
clean_df = df.to_csv('..\clean_data.csv')
```
